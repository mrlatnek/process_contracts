```python
import csv
import scipy.linalg as la
import pandas as pd 
import numpy as np
np.seterr(divide='ignore')
```

```python
years = [str(i) for i in range(1997,2020)]
max_eig_values = []
centralities = []
for year in years:
    print(year)
    sales = []
    with open("NOMINAL_MAKE/NOMINAL_MAKE_"+year+".csv") as csvfile:
        reader = csv.reader(csvfile) # change contents to floats
        for row in reader: # each row is a list
            sales.append(row)

    purchase = []
    with open("NOMINAL_USE/NOMINAL_USE_"+year+".csv") as csvfile:
        reader = csv.reader(csvfile) # change contents to floats
        for row in reader: # each row is a list
            purchase.append(row)
    
    sales = np.array([np.array(i) for i in sales])
    purchase = np.array([np.array(i) for i in purchase])
    
    print(sales.shape)
    print(purchase.shape)
    
    sales = sales[1:,1:]
    purchase = purchase[1:-1, 1:-1]
    
    print(sales.shape)
    print(purchase.shape)
    
    sales = sales.astype(float)
    purchase = purchase.astype(float)
    
    where_are_NaNs = np.isnan(sales)
    sales[where_are_NaNs] = 0
    where_are_NaNs = np.isnan(purchase)
    purchase[where_are_NaNs] = 0
    
    A = 0.25*( sales/sales.sum(axis=1)[:,None] + sales/sales.sum(axis=0)[None, :] + purchase/purchase.sum(axis=1)[:,None] + purchase/purchase.sum(axis=0)[None, :] )
    print(A.shape)
    
    where_are_infs = np.isinf(A)
    A[where_are_infs] = 0
    where_are_NaNs = np.isnan(A)
    A[where_are_NaNs] = 0
    eigen = la.eig(A)
    
    eigen_value, eigen_vector = np.real(eigen[0]), np.real(eigen[1])
    centralities.append(eigen_vector[np.argmax(eigen_value)])
    #pd.DataFrame(eigen_vector[np.argmax(eigen_value)]).to_csv("centrality_"+year+".csv")
    max_eig_values.append([year, np.max(eigen_value)])
```

```python
max_eig_values = np.array([np.array(i) for i in max_eig_values])
pd.DataFrame(max_eig_values).to_csv("max_eigen_values.csv")
pd.DataFrame(np.array(centralities)).to_csv("centralities.csv")
```
